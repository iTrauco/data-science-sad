# 3. Model Development Framework

This section outlines a systematic approach to developing computer vision models for vehicle detection and tracking, with specific focus on model compression and optimization for deployment - key challenges identified by the transportation research team.

## 3.1 Model Architecture Considerations

When approaching vehicle detection and tracking with deployment optimization in mind, model selection requires balancing accuracy with computational efficiency.

### Detection Model Options

For vehicle detection in traffic camera footage, several architectures merit consideration:

- **YOLO Family**: Real-time object detection with good accuracy-speed tradeoff
- **SSD (Single Shot Detector)**: Efficient detection suitable for deployment
- **Faster R-CNN**: Higher accuracy but more computationally intensive
- **MobileNet-based Detectors**: Designed for deployment with built-in efficiency

### Tracking Approaches

Multi-object tracking adds complexity beyond detection:

- **Simple Online Tracking**: Frame-by-frame association based on detection overlap
- **Deep SORT**: Combines detection with appearance features for robust tracking
- **ByteTrack**: Recent approach focusing on utilizing all detection boxes

### Model Compression Considerations

Since the team struggles with model compression, key strategies include:

- **Architecture Selection**: Choose inherently efficient models
- **Quantization**: Reduce model precision for smaller size
- **Pruning**: Remove less important model parameters
- **Knowledge Distillation**: Train smaller models to mimic larger ones

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# conceptual model comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# accuracy vs speed conceptual
ax1.set_xlim(0, 10)
ax1.set_ylim(0, 10)

# model positions (conceptual)
models = [
    (2, 8, 'Faster\nR-CNN', 'green'),
    (6, 6, 'YOLO', 'red'),
    (7.5, 5.5, 'SSD', 'blue'),
    (5, 4.5, 'MobileNet\nSSD', 'orange')
]

for x, y, name, color in models:
    circle = patches.Circle((x, y), 0.5, facecolor=color, alpha=0.6, edgecolor='black')
    ax1.add_patch(circle)
    ax1.text(x, y, name, ha='center', va='center', fontsize=9, fontweight='bold')

ax1.set_xlabel('Inference Speed ', fontsize=12)
ax1.set_ylabel('Detection Accuracy ', fontsize=12)
ax1.set_title('Model Trade-off Space', fontsize=14)

# remove ticks for conceptual nature
ax1.set_xticks([])
ax1.set_yticks([])

# model size conceptual
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)

# size comparison bars (conceptual)
models_list = ['YOLO', 'SSD', 'Faster\nR-CNN', 'MobileNet\nSSD']
x_positions = [2, 4, 6, 8]
original_heights = [6, 4, 8, 2]
compressed_heights = [2, 1.5, 3, 0.8]

for i, (x, orig, comp, model) in enumerate(zip(x_positions, original_heights, compressed_heights, models_list)):
    # original size
    rect1 = patches.Rectangle((x-0.4, 1), 0.35, orig, 
                             facecolor='lightcoral', edgecolor='black')
    ax2.add_patch(rect1)
    # compressed size
    rect2 = patches.Rectangle((x, 1), 0.35, comp, 
                             facecolor='lightblue', edgecolor='black')
    ax2.add_patch(rect2)
    ax2.text(x, 0.5, model, ha='center', fontsize=9)

ax2.text(1, 8, 'Original', color='lightcoral', fontweight='bold')
ax2.text(1, 7.5, 'Compressed', color='lightblue', fontweight='bold')
ax2.set_xlabel('Model Type', fontsize=12)
ax2.set_ylabel('Relative Size ', fontsize=12)
ax2.set_title('Compression Potential', fontsize=14)
ax2.set_xticks([])
ax2.set_yticks([])

plt.suptitle('Model Selection Considerations', fontsize=16)
plt.tight_layout()
plt.show()
```

### Understanding Model Trade-offs

The visualizations above illustrate two critical considerations for model selection:

**Model Trade-off Space (Left)**
- Models are positioned based on their relative speed and accuracy characteristics
- [Faster R-CNN](https://pytorch.org/vision/stable/models/faster_rcnn.html): Higher accuracy but slower inference, suitable when accuracy is paramount
- [YOLO](https://docs.ultralytics.com/): Balanced option offering reasonable accuracy with faster inference
- [SSD](https://pytorch.org/hub/nvidia_deeplearningexamples_ssd/): Single Shot Detector balancing speed and accuracy
- [MobileNet SSD](https://pytorch.org/hub/pytorch_vision_mobilenet_v2/): Optimized for speed and deployment, trading some accuracy for efficiency

**Compression Potential (Right)**
- Shows relative model sizes before and after compression techniques
- Larger models (Faster R-CNN) typically offer more compression opportunity through:
  - [Quantization](https://pytorch.org/docs/stable/quantization.html): Reducing numerical precision
  - [Pruning](https://pytorch.org/tutorials/intermediate/pruning_tutorial.html): Removing less important parameters
  - [Knowledge Distillation](https://pytorch.org/tutorials/beginner/knowledge_distillation_tutorial.html): Training smaller models to mimic larger ones
- Pre-optimized models like [MobileNet](https://pytorch.org/vision/stable/models/mobilenetv2.html) have less compression potential as they're already efficient

These trade-offs directly address the team's challenges with model compression and deployment optimization. The choice depends on specific requirements:
- Real-time processing needs favor faster models
- Accuracy-critical applications may justify slower models
- Deployment constraints might require maximum compression

The visualization above presents the conceptual framework for evaluating our multi-object tracking system across four key dimensions:

- **[FPS (Frames Per Second)](https://docs.ultralytics.com/modes/benchmark/)**: Critical for real-time processing capability assessment

**Tracking Metrics**
- **[MOTA/MOTP](https://motchallenge.net/results/MOT20/)**: Standard MOT Challenge metrics for evaluating multi-object tracking performance

## 4.2 Validation Process Design

For reliable evaluation of the traffic vision system, a structured validation process ensures models perform consistently across different scenarios and camera locations.

### Data Splitting Strategy

**Temporal Consistency**
- Split data by time periods to avoid data leakage
- Ensure training and validation sets don't overlap temporally
- Test on completely separate time periods or camera locations

**Location-Based Validation**
- Reserve specific camera locations for validation
- Test generalization across different highway sections
- Evaluate performance on varied traffic conditions

### Cross-Validation Approach

Given multiple camera locations, implement location-based cross-validation:

1. **Leave-One-Camera-Out**: Train on N-1 cameras, validate on the held-out camera
2. **Temporal Splits**: Within each camera, maintain chronological ordering
3. **Stratified Sampling**: Ensure diverse traffic conditions in each split

### Performance Validation

**Detection Validation**
- Annotate sample videos for ground truth
- Calculate detection metrics on held-out data
- Verify performance across different times of day

**Tracking Validation**
- Manual verification of track consistency
- Evaluation of ID switches at critical points
- Assessment of tracking through occlusions

**Deployment Validation**
- Test compressed models on target hardware
- Measure actual inference times
- Verify memory usage stays within limits

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# validation process flow
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_xlim(0, 12)
ax.set_ylim(0, 10)
ax.axis('off')

# title
ax.text(6, 9.5, 'Validation Process Overview', 
        ha='center', fontsize=16, fontweight='bold')

# camera locations
cameras = ['Camera A', 'Camera B', 'Camera C', 'Camera D', 'Camera E']
camera_y = 7.5

for i, cam in enumerate(cameras):
    x = i * 2 + 2
    rect = patches.Rectangle((x-0.4, camera_y-0.3), 0.8, 0.6,
                           facecolor='lightblue', edgecolor='black')
    ax.add_patch(rect)
    ax.text(x, camera_y, cam, ha='center', va='center', fontsize=9)

# cross-validation illustration
cv_y = 5.5
ax.text(1, cv_y, 'Cross-Validation:', fontsize=11, fontweight='bold')

# show train/val splits
splits = [
    ['Train', 'Train', 'Train', 'Train', 'Val'],
    ['Train', 'Train', 'Train', 'Val', 'Train'],
    ['Train', 'Train', 'Val', 'Train', 'Train']
]

for j, split in enumerate(splits):
    y = cv_y - (j+1) * 0.8
    ax.text(0.5, y, f'Fold {j+1}:', fontsize=9)
    for i, status in enumerate(split):
        x = i * 2 + 2
        color = 'lightgreen' if status == 'Train' else 'lightcoral'
        rect = patches.Rectangle((x-0.35, y-0.25), 0.7, 0.5,
                               facecolor=color, edgecolor='black', alpha=0.7)
        ax.add_patch(rect)

# temporal split illustration
temporal_y = 2
ax.text(1, temporal_y + 0.5, 'Temporal Split:', fontsize=11, fontweight='bold')

# timeline
timeline_x = 2
timeline_width = 8
ax.add_patch(patches.Rectangle((timeline_x, temporal_y-0.2), timeline_width, 0.4,
                             facecolor='lightgray', edgecolor='black'))

# split markers
train_end = timeline_x + timeline_width * 0.7
val_end = timeline_x + timeline_width * 0.85

ax.add_patch(patches.Rectangle((timeline_x, temporal_y-0.2), timeline_width*0.7, 0.4,
                             facecolor='lightgreen', edgecolor='black', alpha=0.7))
ax.add_patch(patches.Rectangle((train_end, temporal_y-0.2), timeline_width*0.15, 0.4,
                             facecolor='lightyellow', edgecolor='black', alpha=0.7))
ax.add_patch(patches.Rectangle((val_end, temporal_y-0.2), timeline_width*0.15, 0.4,
                             facecolor='lightcoral', edgecolor='black', alpha=0.7))

# labels
ax.text(timeline_x + timeline_width*0.35, temporal_y, 'Training', 
        ha='center', va='center', fontsize=9)
ax.text(train_end + timeline_width*0.075, temporal_y, 'Val', 
        ha='center', va='center', fontsize=9)
ax.text(val_end + timeline_width*0.075, temporal_y, 'Test', 
        ha='center', va='center', fontsize=9)

# time arrow
ax.arrow(timeline_x - 0.5, temporal_y - 0.5, timeline_width + 1, 0,
        head_width=0.15, head_length=0.2, fc='gray', ec='gray')
ax.text(6, temporal_y - 0.8, 'Time', ha='center', fontsize=9, style='italic')

plt.tight_layout()
plt.show()
```

### Validation Strategy Explanation

The visualization above illustrates two complementary validation approaches for ensuring robust model performance across different camera locations and time periods:

**Cross-Validation: Leave-One-Camera-Out**
- Each fold trains on data from 4 cameras and validates on the 5th held-out camera
- This approach tests the model's ability to generalize to new locations
- Ensures the system can handle variations in:
  - Camera angles and mounting heights
  - Traffic patterns specific to different highway sections
  - Environmental conditions at different locations
- Provides multiple performance estimates for statistical confidence

**Temporal Split Strategy**
- Chronological splitting prevents data leakage from future to past
- **Training Set (70%)**: Historical data for model development
- **Validation Set (15%)**: Recent data for hyperparameter tuning
- **Test Set (15%)**: Future data for final performance evaluation
- Critical for time-series data where patterns may evolve

**Why Both Approaches Matter**
1. **Spatial Generalization**: Cross-validation ensures models work across different physical locations
2. **Temporal Generalization**: Time-based splits ensure models handle evolving traffic patterns
3. **Realistic Evaluation**: Mimics real deployment where models must work on new cameras and future data

This dual validation strategy directly addresses the challenge of deploying models across multiple highway locations while ensuring consistent performance over time.

## 5.3 System Integration Approach

The integration of HPC and GCP resources requires a coordinated approach to enable efficient video processing and model development workflows.

### Integration Architecture

**Data Flow Integration**
- Camera footage uploads to GCP storage
- HPC workstation retrieves videos for processing
- Results can be stored locally or pushed back to GCP

**Workflow Coordination**
- Download management for 1-hour video segments
- Processing queue for multiple videos
- Result organization and storage

### Processing Pipeline Integration

**Video Processing Workflow**
1. Monitor GCP storage for new video files
2. Download videos to HPC for processing
3. Execute model inference on GPU
4. Store results for analysis

**Model Development Workflow**
1. Access training videos from GCP
2. Train and optimize models on HPC
3. Test compressed models locally
4. Deploy optimized models to inference pipeline

### Operational Considerations

**Data Management**
- Efficient transfer between GCP and HPC
- Local storage management for active processing
- Cleanup of processed videos to manage space

**Pipeline Monitoring**
- Track video processing status
- Monitor resource utilization
- Log processing times and errors
